```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
air_data = pd.read_csv('airbnb-listings.csv', delimiter=';', header=0)
```

```python
air_data.info()
```

```python
air_data['Neighbourhood']
```

```python
#df.columns[df.isna().any()].tolist()
air_data.dropna(axis=1, how='all',inplace=True)
#drops columns that are alll NAN and keeps it like that (inplace)
```

```python
#Find the number of places being in which part of Amsterdam
feq=air_data['Neighbourhood'].value_counts().sort_values(ascending=True)
feq.plot.barh(figsize=(10, 8), color='b', width=1)
plt.title("Number of listings by neighbourhood", fontsize=20)
plt.xlabel('Number of listings', fontsize=12)
plt.show()
```

```python
#Find the types of houses available and general population distribution
feq=air_data['Room Type'].value_counts().sort_values(ascending=True)
feq.plot.barh(figsize=(10, 8), color=["g","b","r"], width=1)
plt.show()
```

```python
air_data.rename(columns={'Property Type':'Property_Type'}, inplace=True)
```

```python
#Types of Properties
feq=air_data['Property_Type'].value_counts().sort_values(ascending=True)
feq.plot.barh(figsize=(10, 8), color=["g"], width=1)
plt.show()
```

```python
air_data.plot(kind="scatter", x="Longitude", y="Latitude",alpha=0.1) 
```

```python
#Accomdates
feq=air_data['Accommodates'].value_counts().sort_values(ascending=True)
feq.plot.bar(figsize=(10, 8), color=["g"], width=1, rot=0)
plt.title("Accommodates (number of people)", fontsize=20)
plt.ylabel('Number of listings', fontsize=12)
plt.xlabel('Accommodates', fontsize=12)
plt.show()
```

```python
feq = air_data[air_data['Accommodates']==2]
feq = feq.groupby('Neighbourhood')['Price'].mean().sort_values(ascending=True)
feq.plot.barh(figsize=(10, 8), color='b', width=1)
plt.title("Average daily price for a 2-persons accommodation", fontsize=20)
plt.xlabel('Average daily price (Euro)', fontsize=12)
plt.ylabel("")
plt.show()
```
